Here is the teanscript for the multiple dot plot

```python

so.Plot(DF-samples, y="result", color="Breed").pair(x=["sample","Ages"]).facet(row="Gender").add(so.Dots(), so.Jitter(0.8)) .theme({"axes.facecolor": "w", "axes.edgecolor": "C0"})
```

Here is the code for the bar plot

```python
colors = ["#FCCF3E", "#F55F20"]
sns.set_palette(sns.color_palette(colors))

fig = sns.barplot(data=DF-P-value, x="Factors", y="-log2(P-value)", hue="Types")
plt.axhline(y=2, color="blue", linestyle="dashed")
plt.savefig("/content/mydrive/MyDrive/fig.png")
```
